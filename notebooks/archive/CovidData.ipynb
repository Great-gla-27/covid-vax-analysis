```python
import pandas as pd
from sqlalchemy import create_engine

# Create SQLite database
engine = create_engine('sqlite:///covid_data.db', echo=False)

# Load your pre-cleaned CSVs
deaths = pd.read_csv('covidDeaths.csv')
vaccinations = pd.read_csv('covidVaccinations.csv')

# Write to SQL tables
deaths.to_sql('covid_deaths', engine, if_exists='replace', index=False)
vaccinations.to_sql('covid_vaccinations', engine, if_exists='replace', index=False)

print("Clean data loaded into SQLite!")
```

```python
#pretty cool rolling count - Too bad it doesnt work with SQlite
rollCnt = pd.read_sql_query("""
    DROP TABLE IF exists #PercentPoulationVaccinated
    CREATETABLE #PercentPoulationVaccinated
    (
    continent nvarchar(255),
    Location nvarchar(255),
    Date datetime,
    Population numeric,
    New_vaccinations numeric,
    RollingPeopleVaccinated numeric
    )


    INSERT into #PercentPoulationVaccinated
    SELECT 
        dea.continent, 
        dea.location, 
        dea.date, 
        dea.population, 
        vac.new_vaccinations,
        SUM(CAST(vac.new_vaccinations AS INTEGER)) 
            OVER (PARTITION BY dea.location ORDER BY dea.date) AS RollingPeopleVaccinated
    FROM covid_deaths dea 
    JOIN covid_vaccinations vac 
        ON dea.location = vac.location 
        AND dea.date = vac.date
    WHERE dea.continent IS NOT NULL
    
    SELECT *, (RollingDosesAdministered * 1.0 / population) AS DosesPerPerson
    FROM #PercentPoulationVaccinated
""", engine)

display(rollCnt)

#rollCnt.to_csv('outputs/rollingCount.csv', index=False)
```

```python
#This yields values like 0.8, 1.5, 3.4 -> meaning doses per person.
#It can exceed 1.0 (100%) without being “wrong,” since people can get multiple doses.
#  rolling doses per capita
rollCTE = pd.read_sql_query("""
    WITH PopvsVac AS (
        SELECT 
            dea.continent, 
            dea.location, 
            dea.date, 
            dea.population, 
            vac.new_vaccinations,
            vac.people_vaccinated,
            vac.people_fully_vaccinated,
            SUM(CAST(vac.new_vaccinations AS INTEGER)) 
                OVER (PARTITION BY dea.location ORDER BY dea.date) AS RollingDosesAdministered
        FROM covid_deaths dea 
        JOIN covid_vaccinations vac 
            ON dea.location = vac.location 
            AND dea.date = vac.date
        WHERE dea.continent IS NOT NULL
    )
    SELECT 
        *,
        (RollingDosesAdministered * 1.0 / population) AS DosesPerPerson,
        (people_vaccinated * 100.0 / population) AS PercentWithAtLeastOneDose,
        (people_fully_vaccinated * 100.0 / population) AS PercentFullyVaccinated
    FROM PopvsVac
    ORDER BY location, date
""", engine)

display(rollCTE)

# Export to CSV
rollCTE.to_csv('outputs/rollingDosesPerCapita.csv', index=False)
```
